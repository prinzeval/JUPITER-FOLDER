```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import html5lib
import urllib.request
```

```python
url = "https://en.wikipedia.org/wiki/List_of_universities_and_colleges_in_Northern_Cyprus"
page =  requests.get(url)
soup = BeautifulSoup(page.text,'html')




```

```python
soup.find_all('table')[0]

```

```python
soup.find_all('table',class_="wikitable sortable mw-collapsible")
```

```python
table = soup.find_all('table')[0]
table
```

```python
Table= table.find_all('th')
Table 
```

```python
Title_data = [line.text.strip() for line in Table]
print(Title_data)
```

```python
df = pd.DataFrame(columns = Title_data)
df
```

```python
column_data = table.find_all('tr')

for rows in column_data[1:]:
   rows_data = (rows.find_all('td'))
   main_rows_data = [rows.text.strip() for rows in rows_data ]

   length = len(df)
   df.loc[length] = main_rows_data
   
```

```python
df
```

```python
df.to_csv(r'C:\Users\Valentine Onuoha\OneDrive\Desktop\webscraping table\allpowers.csv',index = False)
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.text,'html.parser')
soup
```

```python
soup.find_all('table',class_="wikitable sortable")
```

```python
Table = soup.find_all('table')[2]
Table
```

```python
Title = Table.find_all('th')
Title_row  =[line.text.strip() for line in Title]
print(Title_row)
```

```python
df = pd.DataFrame(columns=Title_row)
df
```

```python
row = Table.find_all('tr')
for cols in row[1:]:
    column_data = (cols.find_all('td'))
    my_column_data = [jet.text.strip() for jet in column_data]
    
    length = len(df)
    df.loc[length] = my_column_data
```
